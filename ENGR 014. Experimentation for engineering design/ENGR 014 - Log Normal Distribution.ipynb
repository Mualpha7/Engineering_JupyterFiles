## Lognormal Probability Distribution

### Plot PDF and Generate a Histogram from 1000 Points of Simulated Data

Experiment with different values for mu and sigma

```python
%matplotlib inline
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt
alpha = 1.5 # mean of underlying Normal random variable (scale parameter - first form)
median = np.exp(alpha)    # (scale parameter - second form, used in Python lognormal functions)
beta = 1   # Standard Deviation of underlying Normal random variable (shape parameter)

mean, var, skew, kurt = lognorm.stats(scale=median, s=beta, moments='mvsk')
x = np.linspace(0,30,1000)
plt.plot(x, lognorm.pdf(x,scale=median, s=beta),'r-', lw=3, label='lognormal pdf')
plt.legend(loc='best', frameon=False)
plt.title("Lognormal Distribution",fontsize=20)
plt.xlabel("X",fontsize=20)
plt.ylabel("Density",fontsize=20)
plt.xlim(0,30)
```

### Calculate Probabilities using the CDF

```python
# Be sure to run the cell above to activate the imports
alpha = 1.5  # first form of the scale parameter
median = np.exp(alpha)  # second form of the scale parameter (used in Python)
beta = 1.0   # shape parameter

x1 = 0
x2 = median
F1 = lognorm.cdf(x1, scale=median, s=beta)
F2 = lognorm.cdf(x2, scale=median, s=beta)
P = F2 - F1
print "The probability that x lies between x1 and x2 is %0.4f" % P
```
