## Normal Probability Distribution

### Plot PDF and Generate a Histogram from 1000 Points of Simulated Data

Experiment with different values for mu and sigma

```python
import sys
get_ipython().magic('matplotlib inline')
from scipy.stats import norm
import numpy as np
%matplotlib inline
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
mu = 12.0  # mean (location parameter)
sigma = 2.0   # Standard Deviation (scale parameter)
mean, var, skew, kurt = norm.stats(loc=mu, scale=sigma, moments='mvsk')
x = np.linspace(0,20,100)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma),'r-', lw=3, label='Normal pdf')
data = norm.rvs(loc=mu, scale=sigma, size=1000)
plt.hist(data, normed=True, histtype='stepfilled', alpha=0.2, label='simulated data')
plt.legend(loc='best', frameon=False)
plt.title("Normal (Gaussian) Distribution",fontsize=20)
plt.xlabel("X",fontsize=20)
plt.ylabel("Density",fontsize=20)
```

### Calculate Probabilities using the CDF

```python
# Be sure to run the cell above to activate the imports
mu = 12.0  # Shape parameter
sigma = 2.0   # Scale parameter
x1 = 10
x2 = 15
F1 = norm.cdf(x1, loc=mu, scale=sigma)
F2 = norm.cdf(x2, loc=mu, scale=sigma)
P = F2 - F1
print "The prabability that x lies between x1 and x2 is %0.4f" % P
```
