```python
import pylab
from scipy.ndimage import measurements

#Parameters
L = 100 #Set size of the blood clot
p = 0.4 #Porosity

#We first model the blood clots
r = rand(L,L) #Create random a data set (these are the microbubbles in blue). If we had real data, we could use that instead
z = r < p #We want just as many numbers (i.e. blood clots) as our porosity allows
figure(figsize=(16,5))
subplot(1,3,1)
imshow(z, origin='lower', interpolation='nearest') #Show image
title("Blood clot")
colorbar()

#We then label the clusters, using the scipy command measurements
lw, num = measurements.label(z)
subplot(1,3,2)
b = arange(lw.max() + 1) #We create an array of values from 0 to lw.max() + 1
shuffle(b) #Shuffle it
shuffledLw = b[lw] #Replace all values with values from b
imshow(shuffledLw, origin='lower', interpolation='nearest') #Show image
title("Labeled clusters")
colorbar()

#Finally, calculate the largest area out of all clusters
subplot(1,3,3)
area = measurements.sum(z, lw, index=arange(lw.max() + 1))
areaImg = area[lw]
im3 = imshow(areaImg, origin='lower', interpolation='nearest') #Show image
title("Largest cluster (red)")
colorbar()

show()
```
