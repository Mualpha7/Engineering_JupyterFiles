# Problem 1.5.10

```python
import numpy as np

A = [[1, 2], [0, 3]]
SD = np.linalg.eig(A)
D = SD[0]
S = SD[1]
D_mat = np.diag(D)
S_inv = np.linalg.inv(SD[1])
SDS_inv = np.matmul(S,np.matmul(D_mat,S_inv))
print SDS_inv

A = [[1, 1], [2, 2]]
SD = np.linalg.eig(A)
D = SD[0]
S = SD[1]
D_mat = np.diag(D)
S_inv = np.linalg.inv(SD[1])
SDS_inv = np.matmul(S,np.matmul(D_mat,S_inv))
print SDS_inv
```
