# Problem1.4.11

```python
# Program to plot Quadratic Spline and Cubic Spline

import numpy as np
import matplotlib.pyplot as plt

def quad(x):
    if x < 0:
        return 0
    if x>= 0:
        return x**2/2
    
def cubic(x):
    if x < 0:
        return 0
    if x>= 0:
        return x**3/6
    
q = []
for i in np.arange(-50,50,1):
    q.append(quad(i))
    
c = []
for i in np.arange(-50,50,1):
    c.append(cubic(i))

plt.figure()
plt.plot(np.arange(-50,50,1),q, '-r', label='Quadratic Spline')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
plt.plot(np.arange(-50,50,1),c, '-b', label='Cubic Spline')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
plt.plot(np.arange(-50,50,1),q, '-r', label='Quadratic Spline')
plt.plot(np.arange(-50,50,1),c, '-b', label='Cubic Spline')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
