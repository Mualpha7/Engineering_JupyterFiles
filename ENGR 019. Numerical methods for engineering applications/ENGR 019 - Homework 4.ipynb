### The differential equation $u''(t) + u(t) = 0$ is equivalent to $u''(t) = -u(t)$, for which we have four methods of approximating a solution using computational techniques when the initial conditions are $u(0)=1$ and $u'(0)=0$. Below we use all four methods and reflect on which one is more adequate in this case.

# Problem 1: Forward Euler

```python
# Program that uses Forward Euler to solve and plot u''(t) + u(t) = 0 with initial conditions u(0)=1 and u'(0)=0,
# with three different step sizes.

import numpy as np
import math
import matplotlib.pyplot as plt

#first, we create our matrix for the Forward Euler
def G_f(n, dt): 
    G_f = [[1, dt], [-dt, 1]]
    G_f_prev = G_f
    for i in np.arange(0,n+dt, dt):
        G_f_prev = np.matmul(G_f_prev, G_f)
        count = i
        if count == n:
            return G_f_prev

#these are our initial conditions
u_0 = [1, 0] 

#for each graph we plot, we compute separately the values for the x axis and for the y axis
cx = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    sol = math.cos(i)
    cx.append(sol)
print 'Analytical solution X successful'

cy = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    sol = (-1) * math.sin(i)
    cy.append(sol)
print 'Analytical solution Y successful'
    
f3x = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f3x.append(u_n[0])
print 'Forward Euler X with 1*10^{-3} successful'
    
f3y = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f3y.append(u_n[1])
print 'Forward Euler Y with 1*10^{-3} successful'
    
f2x = []
dt = 10 ** (-2)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f2x.append(u_n[0])
print 'Forward Euler X with 1*10^{-2} successful'
    
f2y = []
dt = 10 ** (-2)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f2y.append(u_n[1])
print 'Forward Euler Y with 1*10^{-2} successful'
    
f1x = []
dt = 10 ** (-1)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f1x.append(u_n[0])
print 'Forward Euler X with 1*10^{-1} successful'    
    
f1y = []
dt = 10 ** (-1)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_f(i, dt), u_0)
    f1y.append(u_n[1])
print 'Forward Euler Y with 1*10^{-1} successful'    

#finally, we plot the four graphs in the same figure using the numbers computed above
plt.figure()
plt.plot(cx, cy, '-r', label='Analytical Solution')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f3x, f3y, '-y', label='Forward Euler with Step Size $1*10^{-3}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f2x, f2y, '-b', label='Forward Euler with Step Size $1*10^{-2}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f1x, f1y, '-g', label='Forward Euler with Step Size $1*10^{-1}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# Problem 2: Backward Euler

```python
# Program that uses Backward Euler to solve and plot u''(t) + u(t) = 0 with initial conditions u(0)=1 and u'(0)=0,
# with three different step sizes.

import numpy as np
import math
import matplotlib.pyplot as plt

#first, we create our matrix for the Backward Euler
def G_b(n, dt):
    G_b = np.linalg.inv([[1, -dt], [dt, 1]])
    G_b_prev = G_b
    for i in np.arange(0,n+dt, dt):
        G_b_prev = np.matmul(G_b_prev, G_b)
        count = i
        if count == n:
            return G_b_prev
        
#these are our initial conditions
u_0 = [1, 0] 

#for each graph we plot, we compute separately the values for the x axis and for the y axis
cx = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    sol = math.cos(i)
    cx.append(sol)
print 'Analytical solution X successful'

cy = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    sol = (-1) * math.sin(i)
    cy.append(sol)
print 'Analytical solution Y successful'
    
f3x = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f3x.append(u_n[0])
print 'Backward Euler X with 1*10^{-3} successful'
    
f3y = []
dt = 10 ** (-3)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f3y.append(u_n[1])
print 'Backward Euler Y with 1*10^{-3} successful'
    
f2x = []
dt = 10 ** (-2)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f2x.append(u_n[0])
print 'Backward Euler X with 1*10^{-2} successful'
    
f2y = []
dt = 10 ** (-2)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f2y.append(u_n[1])
print 'Backward Euler Y with 1*10^{-2} successful'
    
f1x = []
dt = 10 ** (-1)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f1x.append(u_n[0])
print 'Backward Euler X with 1*10^{-1} successful'    
    
f1y = []
dt = 10 ** (-1)
for i in np.arange(0,2*math.pi,dt):
    print 'Step ', i
    u_n = np.matmul(G_b(i, dt), u_0)
    f1y.append(u_n[1])
print 'Backward Euler Y with 1*10^{-1} successful'    
    
#finally, we plot the four graphs in the same figure using the numbers computed above
plt.figure()
plt.plot(cx, cy, '-r', label='Analytical Solution')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f3x, f3y, '-y', label='Backward Euler with Step Size $1*10^{-3}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f2x, f2y, '-b', label='Backward Euler with Step Size $1*10^{-2}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.plot(f1x, f1y, '-g', label='Backward Euler with Step Size $1*10^{-1}$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# Problem 3: Leapfrog Method
