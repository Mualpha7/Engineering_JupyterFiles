```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sys
from mpl_toolkits.mplot3d import Axes3D

def circle(x,y,rad=.015):
    return (np.sqrt(x**2+y**2)<=rad)

M=200
m2 = int(M/2)
sidelen=.2

fx = np.linspace(-sidelen/2,sidelen/2,M)
fy=fx
X,Y=np.meshgrid(fx,fy)
z = circle(X,Y)
Z=np.fft.fftshift(np.fft.fft2(z))
fig = plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X, Y, abs(z), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
fig = plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X, Y, abs(Z), rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
fig = plt.figure()
plt.plot(X[m2],abs(Z[m2,:]))

plt.show()
```
